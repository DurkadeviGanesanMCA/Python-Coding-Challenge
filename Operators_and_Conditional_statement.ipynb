### **Operators**

```python
# Arithmetic Operators
a = 20
b = 30

print(f"Addtion:",a+b)
print(f"Subtraction:",a-b)
print(f"Multiplication:",a*b)
print(f"Division:",a/b)
print(f"Modulus:",a%b)
print(f"Exponent:",a**b)
print(f"Floor Division Operator:",a//b)
```

```python
# Comparison Operators

a = 50
b = 100

print(f"Equal:",a==b)
print(f"Not Equal:",a!=b)
print(f"Greater Than:",a>b)
print(f"Less Than:",a<b)
print(f"Greater Than or Equal to:",a>=b)
print(f"Less Than or Equal to:",a<=b)
```

```python
# Logical Operators

a = 100
b = 200
print(a and b,"And Operator")
print(a or b,"Or Operator")
print(a!=b,"Not Operator")
```

```python
# Membership Operators
text = "I am learing Excel Power BI SQL Python"

print("MemberShip Operator:" not in text)

print("MemberShip Operator:" in text)
```

```python
# Identity Operators # memory
a = [50,25,100]
b = [25,30,40]

c = a
d = b
print(a is b)
print(a is c)
print(a is b)
print(b is d)
print(a is not b)
print(a is not c)
print(a is not b)
print(b is not d)

```

```python
#Assignment Operators

a = 10

a += 5
print(f"Assignment Operators:",a)
a -= 5
print(f"Assignment Operators:",a)
a /= 5
print(f"Assignment Operators:",a)
a *= 5
print(f"Assignment Operators:",a)
a/=5
print(f"Assignment Operators:",a)
a**=5
print(f"Assignment Operators:",a)
a//=5
print(f"Assignment Operators:",a)
a%=5
print(f"Assignment Operators:",a)


```

## if–elif–else **Statement**

```python
# to display the grade of a student based on the marks entered by the user.

marks = int(input("Enter your marks:"))
if marks >= 90:
    print("Grade A")
elif marks >= 80:
    print("Grade B")
elif marks >= 70:
  print("Grade C")
elif marks >= 60:
  print("Grade D")
elif marks >= 50:
  print("Grade E")
else:
  print("Grade F")
```

# **Nested If Condition**

```python
#check the person is eligible for vote or not, chek the citizenship

age = int(input("Enter your age:"))
if age>=18:
  citizenship = input("Enter your citizenship:")
  if citizenship == "Indian":
    print("You are eligible for vote")
    print("Congratulation")
    print("Thank you")
  else:
    print("You are eligible for vote")
    print("Thank you")

else:
    print("You are not eligible for vote")
    print("Thank you")
```

```python
# Case Study Scholarship Eligibility Checker
# Instructions:

# Create variables: student_name, marks, attendance, family_income.

# Use conditional statements (if, elif, else) to:

# Award Full Scholarship if marks ≥ 90 and attendance ≥ 90.
# Award Half Scholarship if marks ≥ 80 or attendance ≥ 85.
# Award Need-Based Scholarship if marks ≥ 70 and family_income < 30000.
# Otherwise, display Not Eligible.
# Display all data (no loops).

Student_name = input("Enter your name:")
marks = int(input("Enter your marks:"))
attendance = int(input("Enter your attendance:"))
family_income = int(input("Enter your family income:"))
if marks >= 90 and attendance >= 90:
    print("Full Scholarship")
elif marks >= 80 or attendance >= 85:
    print("Half Scholarship")
    print("Not Eligible")
elif marks >= 70 and family_income < 30000:
    print("Need-Based Scholarship")
else:
    print("Not Eligible")
```

```python
# Case Study: Bank Loan Eligibility System
# Instructions:
#Create variables: customer_name, age, income, credit_score, existing_loan
#Use conditional statements (if, elif, else) to: • Approve loan if age ≥ 21, income ≥ 30000, credit_score ≥ 700, and no existing loan
# • Conditionally approve loan if credit_score ≥ 600 and income ≥ 40000
# • Otherwise reject the loan application
#Display customer details and loan status using print()

customer_name = input("Enter your name:")
age = int(input("Enter your age:"))
income = int(input("Enter your income:"))
credit_score = int(input("Enter your credit score:"))
existing_loan = input("Enter your existing loan:")
if age >= 21 and income >= 30000 and credit_score >= 700 and existing_loan == "No":
    print("Loan Approved")
    print("Thank you")
elif credit_score >= 600 and income >= 40000:
    print("Loan Approved")
    print("Thank you")
else:
    print("Loan Rejected")
    print("Thank you")


```
