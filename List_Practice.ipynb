```python
# Create List

Cart = ["Keyboard","Mouse","Laptop","Smart Phone"]
print(Cart)
```

```python
# Updating an item (change value using index)
Cart[1] = "Monitor"
print("After update:", Cart)
```

```python
# append() - Add item at end

Cart.append("Washing Machine")
print("After append():", Cart)
```

```python
# insert() - Add item at specific position
Cart.insert(1, "Air Conditioner")
print("After insert():", Cart)
```

```python
# extend() - Add multiple items
Cart.extend(["Refrigerators", "Microwaves","Blenders"])
print("After extend():", Cart)
```

```python
# remove() - Remove first occurrence of value
Cart.remove("Keyboard")
print("After remove():", Cart)
```

```python
# pop() - Remove last item
Cart.pop()
print("After pop():", Cart)
```

```python
# pop() - Remove last item
Cart.pop(2)
print("After pop():", Cart)
```

```python
# index() - Find index of value

print("index():", Cart.index("Mouse"))
```

```python
# count() - Count occurrences of value
print("count():", Cart.count("Mouse"))
```

```python
# sort() - Sort list
Cart = ["Keyboard","Mouse","Laptop","Smart Phone","Washing Machine","Mouse","Laptop"]
Cart.sort()
print("sort():", Cart)
```

```python
# sorted() - Sorted list
Add_cart = sorted(Cart)
print("After Sorted:",Cart)

# reverse() - Reverse list

Add_cart = sorted(Add_cart, reverse=True)

print("After Reversed:", Add_cart)
```

```python
# reverse() - Reverse list

Cart.reverse()
print("reverse():", Cart)
```

```python
# copy() - Copy list
new_Cart = Cart.copy()
print("copy():", new_Cart)
```

```python
# clear() - Remove all items
Clear_Cart = Cart.copy()
Clear_Cart.clear()
print("clear():", Clear_Cart)
```

```python
# List Indexing and Slicing

Cart =["Keyboard","Mouse","Laptop","Smart Phone","Washing Machine","Mouse","Laptop"]
```

```python
# Indexing
print("First item:", Cart[0])
print("Third item:", Cart[2])
print("Last item:", Cart[-1])
```

```python
# Slicing
print("First 3 items:", Cart[0:3])
print("From index 2 to end:", Cart[2:])
print("Reverse list:", Cart[::-1])
```
